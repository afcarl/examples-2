### Import the sample data as a Quilt package
This dataset is daily sales of Noci, a neighborhood gelato shop in Mill Valley, from approximately 2007-2012.

```python
from quilt.data.kmoore import nocisales
```

### Access the sample data as a SQLite Database

```python
import sqlalchemy as sa
eng = sa.create_engine("sqlite:///%s" % nocisales.sqlite())
for row in eng.execute("select * from NociGelato"):        
    print(row)
```
